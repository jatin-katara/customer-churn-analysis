```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.countplot(x='Churn', data=df)
plt.title("customers Churn count")
plt.show()
```

```python
sns.countplot(x='Gender',hue='Churn',data=df)
plt.title("Churn by Gender")
plt.show()
```

```python
sns.boxplot(x='Churn',y='Tenure',data=df)
plt.title("Tenure vs Churn")
plt.show()
```

```python
sns.countplot(x='Subscription Type',hue='Churn',data=df)
plt.title("Churn by Subscription Type")
plt.xticks(rotation=45)
plt.show()
```

```python
df.groupby('Subscription Type')['Churn'].value_counts()
```

```python
sns.boxplot(x='Churn',y='Total Spend',data=df)
plt.title("Total Spend vs Churn")
plt.show()
```
